```python
import pandas as pd

df_results = pd.read_csv('results.csv')
df_status = pd.read_csv('status.csv')
```

```python
dict_status = {}
for row in df_status.itertuples():
  dict_status[row.statusId] = row.status
```

```python
df_results['statusId'] = df_results['statusId'].astype('string')
for idx, row in df_results.iterrows():
  df_results.loc[idx, 'statusId'] = dict_status[int(row['statusId'])]
```
